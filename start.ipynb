```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the file path (relative path as specified)
file_path = "resources/1976-2020-president.csv"

# Load the data
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head(20)
```

```python
# Check the data types and column names to get an overview
data.info()

# Describe numerical columns to understand their range and summary statistics
data.describe()
```

```python
# Check for any missing values
data.isnull().sum()

# Drop rows or handle missing values as appropriate
data.fillna(0, inplace=True)

# Convert any necessary columns to appropriate types, such as date or categorical data
# data['year'] = pd.to_datetime(data['year'], format='%Y')
# Convert 'party_detailed' to categorical type
data['party_detailed'] = data['party_detailed'].astype('category')
```

```python
# Plot voter turnout over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="year", y="totalvotes", hue="party_detailed", marker="o")
plt.title("Voter Turnout by Party from 1976 to 2020")
plt.xlabel("Year")
plt.ylabel("Total Votes")
plt.show()
```

```python
# Calculate the vote share percentage by year and party
data['vote_share'] = data.groupby(['year'])['candidatevotes'].transform(lambda x: x / x.sum() * 100)

# Plot vote share percentage
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="year", y="vote_share", hue="party_detailed", marker="o")
plt.title("Vote Share by Party from 1976 to 2020")
plt.xlabel("Year")
plt.ylabel("Vote Share (%)")
plt.show()
```

```python
# Plot voter turnout over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="year", y="totalvotes", hue="party_detailed", marker="o")
plt.title("Voter Turnout by Party from 1976 to 2020")
plt.xlabel("Year")
plt.ylabel("Total Votes")
plt.show()
```

```python
# Save cleaned data for future use
data.to_csv("resources/cleaned_1976-2020-president.csv", index=False)
```
